```python
import csv
import pandas
```

```python
filename = "../data/BostonHousing.csv"
```

```python
with open(filename, newline='') as csvfile:
    data = csv.reader(csvfile, delimiter=",")
    next(data)
    count = 0
    for row in data:
        num_cols = len(row)
        count += 1
    print("The number of rows is {}".format(count))
    print("The number of cols is {}".format(num_cols))
```
